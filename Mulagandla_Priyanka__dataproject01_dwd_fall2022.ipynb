```python

#created a list to store values in the column 'parent_platforms'
for index, row in df1.iterrows():
    listt = row['parent_platforms']
    listt2 = []
    for element in listt:
      if element['platform']['name'] != None :
        listt2.append(element['platform']['name'])
    df1['parent_platforms'].iloc[index] = listt2
df1
```

**Question 1: What are the top 10 popular game genres?**

```python
#to get top 10 genres and their count of games
top_10_genres = df1['genres'].value_counts(ascending=True).tail(10)
print(top_10_genres)
```

```python
#visualization of top 10 genres and their count of games
fig =px.bar(top_10_genres, title='What are the top 10 popular game genres?', text='value',orientation='h',
       labels={
           'value': "", #removing titles for x and y axes
           'index': ""
       })
fig.update_layout(showlegend=False)
fig.show()
```

Insights:
* Action genre is the most popular genre. 
* 709 games out of 1000 belong to this genre. The 2nd popular is the Adventure genre.


**Question 2: How many games are releasd through years?**

```python
#to see the columns in the dataframe df1
df1.columns
```

```python
#insert a new column 'released year' in the dataframe
df1.insert(3,"released_year", "")
```

```python
#splitting the year from the dates in the column 'released_date' and putting it in the column 'released_year'
for index, row in df1.iterrows():
  df1['released_year'].iloc[index] = row['released_date'].split('-')[0]
df1
```

```python
#created a pivot table to get the release_year and the count of games for each year
pivot = df1.pivot_table(
         index='released_year',
         aggfunc = 'size'
)
pivot
```

```python
#converted pivot table to a dataframe
df2 = pd.DataFrame(pivot)
```

```python
#visualization of the number of games through years
fig = px.line(pivot,
                 title="How many games are releasd through years?",
                 labels={'x':'Years', 'y':'Number of Games'}
                )
fig.update_layout(
    title="How many games are releasd through years?",
    xaxis_title="",
    yaxis_title="Number of Games",
    )
fig.update_layout(showlegend=False)
fig.show()
```

Insights:
* The highest number of games released in the year 2016. 
* Looking at the line chart, we can say that period between 2012 - 2017 has the highest number of releases.

**Question 3: What are the top 8 Gaming Platforms with highest number of game releases?**

```python
#sql query to get number of games in different platforms using group by on the 3 tables 'games', 'platforms', 'devices'
sql = 'select p.name, count(p.pid) as \'count_games\' from games g, platforms p, devices d where g.id = d.id and d.pid = p.pid group by d.pid;'
cursor.execute(sql)
result = cursor.fetchall()

df = pd.DataFrame(result)
df
```

```python
#Visualization of number of games in different platforms
df = df.sort_values(by='count_games', ascending = False).head(8)
fig = px.bar(df, x='name', y='count_games',title='What are the top 8 Gaming Platforms with highest number of game releases?',text='count_games',
             labels={
           'count_games': "",#removing titles for x and y axis
           'name': ""
       }
      )
fig.show()
```

Insights:
* The PC platform has the highest number of games. 
* The Playstation and Xbox stand in 2nd and 3rd places. 
* Looking at the bar chart, we can say that there are least number of games in the mobile phones like IOS and Android.
